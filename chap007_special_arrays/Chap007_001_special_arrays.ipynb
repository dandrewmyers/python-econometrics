### Special Arrays

```python
import numpy as np
```

```python
M, N = 5, 5
```

```python
x = np.ones((M,N)) # M by N array of 1s
x
```

```python
np.ones_like(x)
```

```python
x = np.zeros((M,N)) # M by N array of 0s
x
```

```python
np.zeros_like(x)
```

```python
x = np.empty((M,N)) # M by N empty array
x
```

```python
np.empty_like(x)
```

```python
z = np.eye(N) # identity array
z
```
